I download the stock price data of Nvidia from the beginning of 2022 till the current date. Here, the current date is determined automatically by the today() function from the datetime package, which means I'll get a different (bigger) result every time I run this code on a future date. I also specify the format of the date to be “YYYY-mm-dd,” an important practice to unify the date format.

```python
today = datetime.today().strftime('%Y-%m-%d')
```

```python
data = yf.download("NVDA", start="2022-01-01", end=today)
```

## Visualizing Stock Price Data

First, let's visualize the closing price as a time series plot. As the name suggests, a time series is a sequence of data with a timestamp in each data point. Thus, when plotted on a graph, the horizontal axis indicates the time that flows from left to right, and the vertical axis represents the quantity of interest, that is, the daily closing price. 

if youo look at the schape of the dataframe, it looks ike the Date column has been made the index, evedent in that the number of coulmns is only 6 instead of 7. To fix this, I'll reset the index and name the column appropriately. 

```python
df = data.reset_index().rename(columns={"index":"Date"})
```

```python
df.head()
```

Awesome, now I can draw the graph.

```python
fig = go.Figure(go.Scatter(x=df["Date"], y=df["Close"]))
fig.show()
```

Nvidia is definatly doing the business, should have gotten in around 2022. More analysis to come... For now lets make it a candlestick chart

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
